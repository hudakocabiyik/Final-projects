# 2015-2016 ile 2016-2017 raporlarının karşılaştırılması

```python
sort_rawdata15 = rawdata15.head(10)
sort_rawdata16 = rawdata16.head(10)
sad_rawdata15  = rawdata15.tail(10)
sad_rawdata16  = rawdata16.tail(10)
f, axs = plt.subplots(2,2, figsize = (15,15), constrained_layout = True)

ax=sns.barplot(x="Country", y="Happiness Score", data=sort_rawdata15 , ax = axs[0, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Happiness Score", data=sad_rawdata15, ax = axs[0, 1])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Happiness.Score", data=sort_rawdata16, ax = axs[1, 0])
ax.tick_params(axis = 'x', rotation = 45)
ax=sns.barplot(x="Country",y="Happiness.Score", data=sad_rawdata16, ax = axs[1, 1])
ax.tick_params(axis = 'x', rotation = 45)


axs[0, 0].title.set_text('2015 YILI EN MUTLU 10 ULKE')
axs[0, 1].title.set_text('2015 YILI EN MUTSUZ 10 ULKE')
axs[1, 0].title.set_text('2016 YILI EN MUTLU 10 ULKE')
axs[1, 1].title.set_text('2016 YILI EN MUTSUZ 10 ULKE')
```
